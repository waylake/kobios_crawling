```python
import requests
from bs4 import BeautifulSoup

# URL 정의
url = "https://kobis.or.kr/kobis/business/stat/boxs/findPeriodShowTicketList.do"

# Fake Header 정의
headers = {
    "User-Agent": "Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/91.0.4472.124 Safari/537.36",
    "Referer": "https://kobis.or.kr",
    "Content-Type": "application/x-www-form-urlencoded",
}

# Payload 정의
payload = {
    "CSRFToken": "G1P5qi2QQzdb48cAT4etN5FEw5HKBf3uH2jfE9rWzEI",
    "loadEnd": "0",
    "curTime": "2024-06-07",
    "totScrnCntOrder": "",
    "totScrnCntRatioOrder": "",
    "totDspCntOrder": "",
    "totIssuAmtOrder": "",
    "addTotIssuAmtOrder": "",
    "totIssuCntOrder": "",
    "addTotIssuCntOrder": "",
    "dmlMode": "search",
    "startDate": "2024-06-01",
    "endDate": "2024-06-07",
    "repNationCd": "",
    "wideareaCd": ""
}

# POST 요청 보내기
response = requests.post(url, headers=headers, data=payload)

# BeautifulSoup으로 HTML 파싱
soup = BeautifulSoup(response.text, 'html.parser')

# 영화별 상영점유율 파싱
table_datas = soup.find('tbody', {'id': 'mainTbody'}).find_all('tr')
movie_screen_ratios = []

for row in table_datas:
    movie_name = row.find('td', {'id': 'td_movie'}).get_text(strip=True)
    screen_ratio = row.find('td', {'id': 'td_totDspCntRatio'}).get_text(strip=True)
    movie_screen_ratios.append((movie_name, screen_ratio))

# 결과 출력
for movie, ratio in movie_screen_ratios:
    print(f"영화: {movie}, 상영점유율: {ratio}")
```
